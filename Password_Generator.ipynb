```python
import secrets
import string
import math
def get_password_length():
    while True:
        try:
            length = int(input("Enter desired password length (minimum 8):"))
            if length < 8:
                print("  [!] Password must be at least 8 characters. Try again.")
            elif length > 120:
                print("  [!] Password length cannot exceed 120 characters. Try again.")
            else:
                return length
        except ValueError:
            print("  [!] Invalid input. Please enter a whole number.")
def get_character_pool():
    print("Select character types to include:")
    print("  1 Letters only (a-z, A-Z)")
    print("  2 Letters + Numbers")
    print("  3 Letters + Numbers + Symbols  (recommended)")

    while True:
        choice = input("Enter choice (1/2/3):")
        if choice == "1":
            return string.ascii_letters, "Letters"
        elif choice == "2":
            return string.ascii_letters + string.digits, "Letters + Numbers"
        elif choice == "3":
            return string.ascii_letters + string.digits + string.punctuation, "Letters + Numbers + Symbols"
        else:
            print("  [!] Please enter 1, 2, or 3.")
def generate_password(length, pool):
    guaranteed = []
    if any(c.isdigit() for c in pool):
        guaranteed.append(secrets.choice(string.digits))
    if any(c.isupper() for c in pool):
        guaranteed.append(secrets.choice(string.ascii_uppercase))
    if any(c in string.punctuation for c in pool):
        guaranteed.append(secrets.choice(string.punctuation))
    remaining_length = length - len(guaranteed)
    random_chars = [secrets.choice(pool) for _ in range(remaining_length)]
   
    all_chars = guaranteed + random_chars
    secrets.SystemRandom().shuffle(all_chars)
    return ''.join(all_chars)


def calculate_entropy(length, pool_size):
    return length * math.log2(pool_size)
def classify_strength(entropy):
    if entropy < 40:
        return "WEAK"
    elif entropy < 60:
        return "FAIR"
    elif entropy < 80:
        return "STRONG"
    else:
        return "VERY STRONG"


def main():
    print("=" * 32)
    print("   Random Password Generator")
    print("=" * 32)

    length = get_password_length()
    pool, pool_name = get_character_pool()
    pool_size = len(pool)
    password = generate_password(length, pool)
    entropy = calculate_entropy(length, pool_size)
    strength = classify_strength(entropy)
    
    print("\n" + "=" * 40)
    print(f"  Generated Password : {password}")
    print(f"  Length             : {length} characters")
    print(f"  Character Pool     : {pool_name} ({pool_size} chars)")
    print(f"  Entropy            : {entropy:.2f} bits")
    print(f"  Strength           : {strength}")
    print("=" * 40)

    if entropy < 60:
        print("Increase length or add symbols for stronger security.")
        print("NIST SP 800-63-4 recommends a minimum of 15 characters")
        print("for high-security contexts.")
    else:
        print("Password meets enterprise security standards.")
        
    again = input("Generate another password? (y/n): ").strip().lower()
    if again == 'y':
        print()
        main()


if __name__ == "__main__":
    main()
```
